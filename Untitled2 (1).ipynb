```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# =============================================================================
# Section 1: Data Gathering, Cleaning & Feature Creation
# (Marks Covered: Gather Data, Clean & Prepare, Select Features, Ensure Integrity, Apply Transformations)
# =============================================================================

print("Step 1: Data ko load aur saaf kiya ja raha hai...")

# Gather Relevant Data: CSV file ko load karna.
# Note: This script assumes 'TARP.csv' is in the same folder.
try:
    df = pd.read_csv('TARP.csv')
except FileNotFoundError:
    print("Error: TARP.csv file not found. Please make sure it is uploaded correctly.")
    exit() # Stop the script if the file isn't found

# Clean & Prepare Data (Part 1): Column ke naam se extra space hatana.
df.columns = df.columns.str.strip()

# Select Meaningful Features: Sirf kaam ke columns ko chunna.
relevant_features = ['Soil Moisture', 'Temperature', 'Soil Humidity', 'Time', 'Status']
df_cleaned = df[relevant_features].copy()

# Ensure Data Integrity (Part 1): Missing values waali rows ko hatana.
df_cleaned.dropna(inplace=True)

# Ensure Data Integrity (Part 2): Duplicate rows ko hatana.
df_cleaned.drop_duplicates(inplace=True)

# Ensure Data Integrity (Part 3): Data types ko theek karna taaki numbers text na bane rahein.
for col in ['Soil Moisture', 'Temperature', 'Soil Humidity', 'Time']:
    df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')
df_cleaned.dropna(inplace=True)

print("Data saaf ho chuka hai.")

# Apply Transformations for Better Representation: Ek nayi, meaningful feature banana.
# Hum 'Soil Moisture' ke basis par paani ke level ko categorize kar rahe hain.
print("Step 2: Nayi feature 'Moisture Level' banayi ja rahi hai...")
conditions = [
    (df_cleaned['Soil Moisture'] < 30),
    (df_cleaned['Soil Moisture'] >= 30) & (df_cleaned['Soil Moisture'] < 60),
    (df_cleaned['Soil Moisture'] >= 60)
]
levels = ['Low', 'Medium', 'High']
df_cleaned['Moisture Level'] = np.select(conditions, levels, default='Unknown')
print("Nayi feature bann gayi.")

# Final Step for Power BI: Saaf kiye hue data ko ek nayi CSV file mein save karna.
df_cleaned.to_csv('cleaned_for_powerbi.csv', index=False)
print("Power BI ke liye 'cleaned_for_powerbi.csv' file taiyaar hai.")
print("-" * 50)
```

```python
# =============================================================================
# Section 2: Descriptive Insights & Visualization
# (Marks Covered: Generate Insights, Discover Patterns, Build Charts, Select Visuals, Aesthetics)
# =============================================================================

print("Step 3: Data se insights nikaale ja rahe hain...")

# Generate Descriptive Insights: Data ki summary (mean, std, etc.) dekhna.
print("Data ki Descriptive Summary:")
print(df_cleaned.describe())
print("-" * 50)

print("Step 4: Presentation ke liye charts banaye ja rahe hain...")

# Set style for better aesthetics
sns.set_style("whitegrid")

# Chart 1: Build Initial Charts - Status ka distribution dekhna.
plt.figure(figsize=(8, 6))
sns.countplot(x='Status', data=df_cleaned, palette='viridis')
plt.title('Watering Status ka Distribution (ON/OFF)', fontsize=16)
plt.xlabel('Status', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.savefig('1_status_distribution.png')
print("Chart 1 (Status Distribution) save ho gaya.")

# Chart 2: Discover Patterns - Alag-alag sensor readings ki frequency dekhna.
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
df_cleaned['Soil Moisture'].hist(ax=axes[0], color='skyblue', edgecolor='black')
axes[0].set_title('Soil Moisture')
df_cleaned['Temperature'].hist(ax=axes[1], color='salmon', edgecolor='black')
axes[1].set_title('Temperature')
df_cleaned['Soil Humidity'].hist(ax=axes[2], color='lightgreen', edgecolor='black')
axes[2].set_title('Soil Humidity')
plt.suptitle('Sensor Readings ke Histograms', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('2_histograms.png')
print("Chart 2 (Histograms) save ho gaya.")

# Chart 3: Select Visualization to Convey Insights - ON/OFF status ke hisaab se Soil Moisture ko compare karna.
plt.figure(figsize=(8, 6))
sns.boxplot(x='Status', y='Soil Moisture', data=df_cleaned, palette='pastel')
plt.title('Watering Status ke Hisaab se Soil Moisture', fontsize=16)
plt.xlabel('Status', fontsize=12)
plt.ylabel('Soil Moisture', fontsize=12)
plt.savefig('3_boxplot_moisture_vs_status.png')
print("Chart 3 (Box Plot) save ho gaya.")

# Chart 4: Discover Hidden Patterns - Sabhi numerical features ka aapas mein relation dekhna.
plt.figure(figsize=(10, 8))
correlation_matrix = df_cleaned[['Soil Moisture', 'Temperature', 'Soil Humidity', 'Time']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Features ka Correlation Heatmap', fontsize=16)
plt.savefig('4_correlation_heatmap.png')
print("Chart 4 (Correlation Heatmap) save ho gaya.")

print("\nSaare kaam poore ho gaye! ðŸš€")
```
